### Bulduğumuz medyanları karşılaştırma

```python
median_1=data_afrika_kıtası["C2O EMİSYON ORANI"].median()
median_2=data_AB["C2O EMİSYON ORANI"].median()

print(median_1)
print(median_2)


if median_1 == median_2:
    print("data_afrika_kıtası veri setinedeki C2O emisyon oranı ve data_AB veri setindeki C2O emiyonu eşittir.")
elif median_1 > median_2:
    print("data_afrika_kıtası veri setindeki C2O emisyon oranı daha fazla .")
else:
    print("data_AB veri setindeki C2O emisyon oranı daha fazla.")
```
